###### 개별 종목의 가격 데이터 가져오기
    - 개별 종목의 가격을 가져오고, 개별 종목의 종가를 데이터 프레임에 컬럼으로 추가

```python
import pandas as pd
import FinanceDataReader as fdr

# 빈 데이터프레임 생성 ******
med = pd.DataFrame()
# 첫번째 변수idx 에 인덱스를 받고, row는 'Symbol'부터 'Region'열의 행에 하나씩 접근하여 출력.(일반의 loop보다 성능 good)
for idx, row in med_sec.iterrows():
    code, name = row['Symbol'], row['Name']
    print(code, name)

    # 개별 종목의 가격 가져오기     
    df = fdr.DataReader(code, '2019-01-01', '2019-12-31')
                            
    # 가격 데이터의 종가를 컬럼으로 추가하기
    med[name] = df['Close']
```

```python
med = med.dropna(axis=1)
med
```

###### 수익률 계산

```python
acc_rets = med / med.iloc[0] - 1.0
acc_rets
```

```python
returns = acc_rets.iloc[-1]
returns.sort_values(ascending=False)
```

```python
returns.mean()
```

###### 다양한 기간에 대한 수익률

```python
# +a특정기간 지정 _ text

df = med['2019-12-01':'2019-12-30'] 
acc_rets = df / df.iloc[0] - 1.0
acc_rets.iloc[-1]
```

- '2019-12-30' 시점을 기준으로 과거 5일, 10일, 20일, 60일, 120일, 240일 각각 수익률을 구하기.

```python
from datetime import datetime, timedelta

the_day = datetime(2019, 12, 30)

for days in [5, 10, 20, 60, 120, 240]:
  start = the_day - timedelta(days)
  end = the_day
  print(start, '~', end)
```

```python
the_day = datetime(2019, 12, 30)
row_dict = {}
for days in [5, 10, 20, 60, 120, 240]:
  start = the_day - timedelta(days)
  end = the_day

  df = med[start:end] 
  acc_rets = df / df.iloc[0] - 1.0
  row_dict[days] = acc_rets.iloc[-1]
```

```python
df_rets = pd.DataFrame(row_dict)
df_rets
```

```python
df_rets.mean()
```

```python
def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color

def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

df_rets.style.\
    applymap(color_negative_red).\
    apply(highlight_max)
```
