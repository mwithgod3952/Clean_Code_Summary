## 수정주가. Adjusted Price
### 요약
    - 수정주가란 총 발행 주식수의 변화를 반역한 가격
    - 벡테스트나 수익률 분석을 위해서 수정주가를 사용
### 개요
    - 수정주가란 총 발행 주식수의 변화를 반영한 가격을 말한다. 액면분할, 액면병합, 유상증자, 무상증자와 같이 총 발행주식
    수의 변화가 발생하면 가격의 변화가 생긴다. 
    예를 들어, 어떤 회사가 액면분할을 통해 1주를 2개로 쪼갠다고 하면, 주식수가 2배로 늘어나고 주식의 가격의 0.5배로 낮아
    지며, 시가총액은 동일하게 유지된다.
    
> **요약 : 시가총액은 유지, but 주식 가격의 분할된 주식수 n에 대하여, n만큼 주시가격이 하락하는 현상**
